```python
df.drop(['area_type','availability','society','balcony'],axis=1,inplace=True)
```

```python
df
```

```python
df['location']=df['location'].fillna(df['location'].mode()[0])
df['size']=df['size'].fillna(df['size'].mode()[0])
df['bath']=df['bath'].fillna(df['bath'].mean())
```

```python
df.isna().sum()
```

```python
df['size']=df['size'].str.replace('BHK','')
df['size']=df['size'].str.replace('Bedroom','')
df['size']=df['size'].str.replace('RK','')
df
```

```python
df['size']=df['size'].astype(int)
```

```python
df.dtypes
```

```python
df['total_sqft'].unique()
```

```python
def convertRange(x):
  temp=x.split('-')
  if len(temp)==2:
    return (float(temp[0])+float(temp[1]))/2
  try:
      return float(x)
  except:
       return None
```

```python
df['total_sqft']=df['total_sqft'].apply(convertRange)
df.head()
```

```python
df['price_per_sqft']=df['price']*100000/df['total_sqft']
```

```python
df.head()
```

```python
df['total_sqft'].unique()
```

```python
df.isna().sum()
```

```python
df['total_sqft']=df['total_sqft'].fillna(df['total_sqft'].mean())
```

```python
df1=pd.get_dummies(df['location'],drop_first=True)
```

```python
df1.head()
```

```python
dfe=pd.concat([df1,df],axis=1)
dfe
```

```python
dfe.drop(['location','price_per_sqft'],axis=1,inplace=True)
dfe.head()
```

```python
x=dfe.iloc[:,:-1].values
y=dfe.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
y_train
y_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df2=pd.DataFrame({'act_value':y_test,'pred_value':y_pred,'diff':y_test-y_pred})
df2
```

```python
from sklearn.metrics import mean_absolute_percentage_error
print('MAPE: ',mean_absolute_percentage_error(y_test,y_pred))
```

```python
from sklearn.metrics import r2_score
print('R2_SCORE: ',r2_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
rfg=RandomForestRegressor()
rfg.fit(x_train,y_train)
y_pred2=rfg.predict(x_test)
y_pred2
```

```python
df3=pd.DataFrame({'act_value':y_test,'pred_value':y_pred2,'diff':y_test-y_pred2})
df3
```

```python
from sklearn.metrics import mean_absolute_percentage_error
print('MAPE: ',mean_absolute_percentage_error(y_test,y_pred2))
```

```python
from sklearn.metrics import r2_score
print('R2 SCORE: ',r2_score(y_test,y_pred2))
```

```python
from xgboost import XGBRegressor
xgb=XGBRegressor()
xgb.fit(x_train,y_train)
y_pred3=xgb.predict(x_test)
y_pred3
```
