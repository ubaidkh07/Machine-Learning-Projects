**3)Encoding columns using get dummies**

```python
df1=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type','Brand']],drop_first=True)
df1
```

**4)Concatination two tables**

```python
dfe=pd.concat([df,df1],axis=1)
dfe
```

**5)Remove unwanted columns**

```python
dfe=dfe.drop(['Unnamed: 0','Name','Location','Fuel_Type','Transmission','Owner_Type','New_Price','Fuel_Type_Electric','Brand'],axis=1)
dfe
```

**6)Remove string value column and replace the value using ' '**

```python
dfe['Mileage']=dfe['Mileage'].str.replace('km/kg','')
dfe['Mileage']=dfe['Mileage'].str.replace('kmpl','')
dfe['Engine']=dfe['Engine'].str.replace('CC','')
dfe['Power']=dfe['Power'].str.replace('bhp','')
dfe['Mileage']=dfe['Mileage'].str.replace('null','0')
dfe['Engine']=dfe['Engine'].str.replace('null','0')
dfe['Power']=dfe['Power'].str.replace('null','0')
```

```python
dfe.dtypes
```

```python
dfe.isna().sum()
```

**7)Convert data type (object to float)**

```python
#datatype convertion
#mileage ==> float
#engine  ==> float
#power ==> float
dfe['Mileage']=dfe['Mileage'].astype(float)
dfe['Engine']=dfe['Engine'].astype(float)
dfe['Power']=dfe['Power'].astype(float)
```

```python
dfe.dtypes
```

```python
dfe.loc[dfe.Mileage==0,'Mileage']=np.NaN
dfe.loc[dfe.Power==0,'Power']=np.NaN
dfe.loc[dfe.Engine==0,'Engine']=np.NaN
```

```python
dfe.isna().sum()
```

**8)Data cleaning**

```python
x3=dfe['Mileage'].mean()
dfe['Mileage'].fillna(x3,inplace=True)
y7=dfe['Engine'].mean()
dfe['Engine'].fillna(y7,inplace=True)
z=dfe['Power'].mean()
dfe['Power'].fillna(z,inplace=True)
a=dfe['Seats'].mode()[0]
dfe['Seats'].fillna(a,inplace=True)
dfe.isna().sum()
```

**9)Separate as X&Y**

```python
x=dfe.drop(['Price'],axis=1)
y=dfe['Price']
```
