```python
x=df.drop(['name','status'],axis=1)
x
```

```python
y=df['status']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
y_train
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
matr=confusion_matrix(y_pred,y_test)
matr
score=accuracy_score(y_test,y_pred)
score
```

```python
knn1=KNeighborsClassifier()
```

```python
param={'n_neighbors':[3,5,7,9],'weights':['uniform','distance']}
clf=GridSearchCV(knn1,param,cv=10,scoring='accuracy')
clf.fit(x_train,y_train)
```

```python
print(clf.best_params_)
```

```python
knn2=KNeighborsClassifier(n_neighbors=3,weights='distance')
knn2.fit(x_train,y_train)
y_pred1=knn2.predict(x_test)
y_pred1
```

```python
print('accuracy_score',accuracy_score(y_test,y_pred1))
```

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.svm import SVC
model=BernoulliNB()
model3=SVC()
lst=[model,model3]
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
for i in lst:
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print(i)
  print(confusion_matrix(y_test,y_pred))
  print("accuracy_score:",accuracy_score(y_test,y_pred))
  print("classification_report:",classification_report(y_test,y_pred))
```
