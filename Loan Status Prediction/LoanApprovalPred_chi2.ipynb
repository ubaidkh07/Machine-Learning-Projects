```python
col=['Gender','Married','Dependents','Self_Employed']
for i in col:
  df[i]=df[i].fillna(df[i].mode()[0])
```

```python
col1=['LoanAmount','Loan_Amount_Term','Credit_History']
for i in col1:
  df[i]=df[i].fillna(df[i].mean())
```

```python
df.isna().sum()
```

```python
col3=['Loan_ID','Gender','Married','Dependents','Education','Self_Employed','Property_Area']
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in col3:
  df[i]=le.fit_transform(df[i])
df
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.feature_selection import chi2
score=chi2(x,y)
score
```

```python
f_value=pd.Series(score[0],index=x.columns)
f_value.sort_values(ascending=False)
```

```python
s_value=pd.Series(score[1],index=x.columns)
s_value.sort_values(ascending=False)
```

```python
df.drop(['Loan_ID','Married','Self_Employed','Gender','Property_Area','Dependents','Education'],axis=1,inplace=True)
df.head()
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
y_train
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.svm import SVC
knn=KNeighborsClassifier(n_neighbors=7)
model=BernoulliNB()
model3=SVC()
lst=[knn,model,model3]
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
for i in lst:
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print(i)
  print(confusion_matrix(y_test,y_pred))
  print("accuracy_score:",accuracy_score(y_test,y_pred))
  print("classification_report:",classification_report(y_test,y_pred))
```
